# Pandas Group By

```python
import pandas as pd
# Create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
dataFrame = pd.DataFrame(data)
```

```python
dataFrame
```

```python
byCompany = dataFrame.groupby('Company')
```

```python
byCompany
```

```python
byCompany.median(numeric_only=True)
```

```python
byCompany.sum(numeric_only=True)
```

```python
byCompany.sum(numeric_only=True).loc['MSFT']
```

```python
dataFrame
```

```python
dataFrame.groupby('Company').describe().transpose()
```
