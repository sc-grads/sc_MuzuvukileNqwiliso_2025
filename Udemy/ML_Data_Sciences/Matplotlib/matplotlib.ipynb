```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel("Y Label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

```python
# 00

fig = plt.figure()
# THIS IS THE CONTAINER OF THE PLOTS

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Plot Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.3,0.5,0.5])
axes1.plot(x,y)
axes2.plot(x,y)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols= 2)
# type(axes)
# type(fig)

axes[0].plot(x,y,'b')
axes[0].set_title('First Plt')
axes[1].plot(y,x,'r')
axes[1].set_title('Second Plt')

plt.tight_layout()
```

# **Figure Size and DPI**

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,  figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(y,x*3, label="X Squared")
axes.plot(y,x**3, label="X Cubed")

axes.legend(loc=1)
# THIS HAS SOME LOCATION WHERE YOU WANT TO PLACE YOUR LABELS OR LEGENDS
plt.savefig('plot2.png',dpi=300)
```

# **Plot Appearance**

```python
 fig = plt.figure()
 axes = fig.add_axes([0,0,1,1])
 axes.plot(x,y,color='red',lw = 3,alpha = 0.5,ls='--',marker='o',ms=10)
 axes.set_xlim(0,1)
 axes.set_ylim(0,2)
```
